Hàm tham chiếu giữa các cấp với nhau

```python
class Address_Obj(object):
  pass
def create_dict(dict_df):
    add_dicts = Address_Obj()

    # Tinh, Thanh pho -> Quan, Huyen, Thi Xa
    tp_qh_dict = {}
    for tp in dict_df['Tỉnh Thành Phố']:
        if tp not in tp_qh_dict:
            tp_qh_dict[tp] = []
        else:
            continue
        for qh in dict_df[dict_df['Tỉnh Thành Phố'] == tp]['Quận Huyện']:
            if qh not in tp_qh_dict[tp]:
                tp_qh_dict[tp].append(qh)

    # Quan, Huyen, Thi Xa -> Xa, Phuong, Thi tran
    qh_xp_dict = {}
    for qh in dict_df['Quận Huyện']:
        if qh not in qh_xp_dict:
            qh_xp_dict[qh] = []
        else:
            continue
        for xp in dict_df[dict_df['Quận Huyện'] == qh]['Phường Xã']:
            if xp not in qh_xp_dict[qh]:
                qh_xp_dict[qh].append(xp)

    add_dicts.tp_qh_dict = tp_qh_dict
    add_dicts.qh_xp_dict = qh_xp_dict

    return add_dicts
```

```python
add_dicts = create_dict(dict_data)
```

Hàm chuẩn hoá các từ viêt tắt

```python
def add_norm(data, chuanhoa):
    for i in range(len(data)):
        for full, abb in chuanhoa.items():
            for j in range(len(abb)):
                if str(abb[j] + ' ') in (data.loc[i, 'Address_ch'] + ' '):
                    data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(abb[j], full)
                data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(',', '')

    return data
```

Hàm tách địa chỉ

```python
# Function to fill city, district, ...
def city_district(data, dict_data, ref_address):

    # Find city, district,...
    for i in range(len(data)):
        for key, values in dict_data.items():
            if (key + ' ') in (data.loc[i, 'Address_ch'][-16:] + ' '):
                if data.loc[i, 't_check'] != 1:
                    data.loc[i, 't_check'] = 1
                    data.loc[i, 'tinh'] = key
                    data.loc[i, 'tinh_cat'] = ref_address[ref_address['Tỉnh Thành Phố']==key]['Cấp Tỉnh/Thành Phố'].values[0]
                    data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(key, '')
            # Check district
            for value in values:
                copy_value = value
                if len(str(value)) == 1 or len(str(value)) == 2:
                    value = str('quận ' + str(value))
                if (value + ' ') in (data.loc[i, 'Address_ch'][-18:] + ' '):
                    if data.loc[i, 'h_check'] != 1:
                        if value == '2' or value == 'thủ đức':
                            data.loc[i, 'qh'] = 'thành phố thủ đức'
                        else:
                            data.loc[i, 'qh'] = value
                        if len(ref_address[(ref_address['Tỉnh Thành Phố']==key) & (ref_address['Quận Huyện']==copy_value)]['Cấp Quận/Huyện'].values) != 0:
                            data.loc[i, 'qh_cat'] = ref_address[(ref_address['Tỉnh Thành Phố']==key) & (ref_address['Quận Huyện']==copy_value)]['Cấp Quận/Huyện'].values[0]
                        data.loc[i, 'h_check'] = 1
                        data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(value, '')
                    # Fill city if city not found
                    if data.loc[i, 't_check'] != 1:
                        data.loc[i, 'tinh'] = key
                        data.loc[i, 'tinh_cat'] = ref_address[ref_address['Tỉnh Thành Phố']==key]['Cấp Tỉnh/Thành Phố'].values[0]
                        data.loc[i, 't_check'] = 1
                        data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(key, '')

    return data

# Function to fill ward, ...
def district_ward(data, dict_data, ref_address):

    # Check district
    for i in range(len(data)):
        if data.loc[i, 'h_check'] == 1:
            for key_1, values_1 in dict_data.items():
                if data.loc[i, 'qh'] == key_1:
                    for value_1 in values_1:
                        # Fill ward, ...
                        copy_value_1 = value_1
                        if len(str(value_1)) == 1 or len(str(value_1)) == 2:
                            value_1 = str('phường ' + str(value_1))
                        if (value_1 + ' ') in (data.loc[i, 'Address_ch'] + ' '):
                            data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(value_1, '')
                            data.loc[i, 'px'] = value_1
                            if len(ref_address[(ref_address['Quận Huyện']==key_1) & (ref_address['Phường Xã']==copy_value_1)]['Cấp Phường/Xã'].values) != 0:
                                data.loc[i, 'px_cat'] = ref_address[(ref_address['Quận Huyện']==key_1) & (ref_address['Phường Xã']==copy_value_1)]['Cấp Phường/Xã'].values[0]

    return data

# Function to fill address
def add_proc_1(data, add_dicts, ref_address):

    # Fill city, district, ...
    city_district(data, add_dicts.tp_qh_dict, ref_address)

    # Fill ward, ...
    district_ward(data, add_dicts.qh_xp_dict, ref_address)

    return data

# Function to remove redundant words after fill
def add_proc_2(data):

    # Redundant words
    redundants = ['mặt đường', "đường lớn", 'thị xã', "thị trấn", 'thành phố', "đường", '-', '( )', 'tt', "trung tâm",
                 "phường", "huyện", "tỉnh", "tx", "tp", "quận", "xã", "quận"]

    for i in range(len(data)):
        if not data.loc[i, 'tinh'] is None:
            for j in redundants:
                if j in data.loc[i, 'Address_ch']:
                    data.loc[i, 'Address_ch'] = data.loc[i, 'Address_ch'].replace(j, "")
            data.loc[i, 'Address_ch'] = re.sub("\s\s+", " ", data.loc[i, 'Address_ch'])

    return data

# Function to fill and update address
def update_entity_address(data, add_dicts, ref_address, abbreviations):
    # Add checking city, district, ... columns
    data['t_check'] = 0
    data['h_check'] = 0

    # Update address_ch
    data['Address_ch'] = data['Address'].replace("_", " ")

    # Update other address
    data = add_norm(data, abbreviations)
    data = add_proc_1(data, add_dicts, ref_address)
    data = add_proc_2(data)

    return data
```
